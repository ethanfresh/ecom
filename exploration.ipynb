```python
import pandas as pd
import requests
import time
import numpy as np
from concurrent.futures import ThreadPoolExecutor, as_completed

df["DATE_DAY"] = pd.to_datetime(df["DATE_DAY"]).dt.date

# 2) Pick monetary columns to convert
money_cols = [
    c for c in df.columns
    if c.endswith("_SPEND") or "PRICE" in c or "DISCOUNT" in c
]
print(f"Selected {len(money_cols)} monetary columns for conversion.")

# 3) Build FX rate lookup: (date, currency) -> USD rate
pairs = df[["DATE_DAY", "CURRENCY_CODE"]].drop_duplicates()
fx = {}
print(f"Preparing to fetch FX rates for {len(pairs)} unique (date, currency) pairs.")

# USD rows require no API call
usd_pairs = pairs[pairs["CURRENCY_CODE"] == "USD"]
for d, c in usd_pairs.itertuples(index=False):
    fx[(d, c)] = 1.0
print(f"Mapped {len(usd_pairs)} USD pairs without API calls.")

non_usd_pairs = [
    (d, c)
    for d, c in pairs.itertuples(index=False)
    if c != "USD" and pd.notna(c)
]
print(f"Calling FX API for {len(non_usd_pairs)} non-USD pairs in parallel...")

def fetch_rate(d, c, retries=3):
    url = f"https://api.frankfurter.app/{d}?from={c}&to=USD"
    for attempt in range(1, retries + 1):
        try:
            r = requests.get(url, timeout=20)
            r.raise_for_status()
            rate = r.json().get("rates", {}).get("USD")
            if rate is None:
                return d, c, None, url, "missing_rate"
            return d, c, float(rate), url, None
        except requests.HTTPError as e:
            status_code = e.response.status_code if e.response is not None else None
            # Frankfurter returns 404 for unsupported currencies (for example CLP).
            if status_code == 404:
                return d, c, None, url, "unsupported_or_missing_on_source"
            if attempt == retries:
                return d, c, None, url, f"http_{status_code}"
            time.sleep(0.5 * attempt)
        except Exception as e:
            if attempt == retries:
                return d, c, None, url, f"exception: {e}"
            time.sleep(0.5 * attempt)

failed_pairs = []
max_workers = min(16, max(1, len(non_usd_pairs)))
if non_usd_pairs:
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = {executor.submit(fetch_rate, d, c): (d, c) for d, c in non_usd_pairs}
        total = len(futures)
        for i, future in enumerate(as_completed(futures), start=1):
            d, c = futures[future]
            d, c, rate, url, error_reason = future.result()
            if rate is None:
                failed_pairs.append((d, c, error_reason, url))
                print(f"[{i}/{total}] WARNING: no FX rate for {c} on {d} ({error_reason})")
                continue
            fx[(d, c)] = rate
            print(f"[{i}/{total}] Retrieved rate: 1 {c} = {rate} USD ({d})")

if failed_pairs:
    print(f"WARNING: Failed to resolve {len(failed_pairs)} FX pairs. Those rows will remain unconverted.")
    preview = failed_pairs[:5]
    for d, c, reason, _ in preview:
        print(f" - {d} {c}: {reason}")

# 4) Apply conversion
print("Applying FX rates to dataframe...")
pair_index = pd.MultiIndex.from_frame(df[["DATE_DAY", "CURRENCY_CODE"]])
df["FX_TO_USD"] = pair_index.map(fx)
print("FX_TO_USD column created.")

missing_fx_rows = int(df["FX_TO_USD"].isna().sum())
if missing_fx_rows:
    print(f"WARNING: {missing_fx_rows} rows have no FX rate and were not converted.")

# Vectorized conversion is faster than per-column loops
usd_cols = [f"{col}_USD" for col in money_cols]
df[usd_cols] = df[money_cols].mul(df["FX_TO_USD"], axis=0)
print(f"Created {len(usd_cols)} *_USD columns.")

# Overwrite originals only for rows with a valid FX rate
valid_mask = df["FX_TO_USD"].notna()
df.loc[valid_mask, money_cols] = df.loc[valid_mask, money_cols].mul(
    df.loc[valid_mask, "FX_TO_USD"], axis=0
)
print("Original monetary columns converted to USD values where FX was available.")

# Only mark rows as USD if conversion happened
df["CURRENCY_CODE"] = np.where(valid_mask, "USD", df["CURRENCY_CODE"])
print("Currency normalization complete for successfully converted rows.")
```

```python
import numpy as np

# Add core time controls to reduce bias in ROI modeling.
df["DATE_DAY"] = pd.to_datetime(df["DATE_DAY"])

df["day_of_week"] = df["DATE_DAY"].dt.dayofweek  # Monday=0, Sunday=6
df["month"] = df["DATE_DAY"].dt.month

# Monotonic time index across the dataset.
df = df.sort_values("DATE_DAY").reset_index(drop=True)
df["trend"] = np.arange(len(df), dtype=int)

print("Created control features: day_of_week, month, trend")
df[["DATE_DAY", "day_of_week", "month", "trend"]].head()
```

```python
# Final clean modeling table.
required_model_cols = [
    "DATE_DAY",
    "ORGANISATION_ID",
    "GOOGLE_TOTAL_SPEND",
    "META_TOTAL_SPEND",
    "TIKTOK_SPEND",
    "revenue",
    "day_of_week",
    "month",
    "trend",
]

missing_model_cols = [c for c in required_model_cols if c not in df.columns]
if missing_model_cols:
    raise KeyError(f"Missing required modeling columns: {missing_model_cols}")

model_df = df[required_model_cols].rename(
    columns={
        "DATE_DAY": "date",
        "ORGANISATION_ID": "brand_id",
        "GOOGLE_TOTAL_SPEND": "spend_google",
        "META_TOTAL_SPEND": "spend_meta",
        "TIKTOK_SPEND": "spend_tiktok",
        "trend": "time_index",
    }
).copy()

# Treat missing TikTok spend as zero spend.
model_df["spend_tiktok"] = model_df["spend_tiktok"].fillna(0.0)

# Keep a consistent order for modeling pipelines.
model_df = model_df[[
    "date",
    "brand_id",
    "spend_google",
    "spend_meta",
    "spend_tiktok",
    "revenue",
    "day_of_week",
    "month",
    "time_index",
]].sort_values(["date", "brand_id"]).reset_index(drop=True)

print("Final modeling table created.")
print(f"Shape: {model_df.shape[0]:,} rows x {model_df.shape[1]} columns")
model_df
```

```python
# Marginal ROI from a log-log model.
channels = ["spend_google", "spend_meta", "spend_tiktok"]

# Log-log requires strictly positive values.
fit_df = model_df[["revenue"] + channels].copy()
valid_fit = (fit_df["revenue"] > 0)
for c in channels:
    valid_fit &= (fit_df[c] > 0)

fit_df = fit_df.loc[valid_fit].copy()
dropped = len(model_df) - len(fit_df)
print(f"Rows used for log-log fit: {len(fit_df):,} | Dropped (non-positive revenue/spend): {dropped:,}")

if len(fit_df) < 30:
    raise ValueError("Too few rows for stable log-log fit after filtering positives.")

# OLS via numpy (no extra dependency).
X = np.column_stack([np.ones(len(fit_df))] + [np.log(fit_df[c].values) for c in channels])
y_log = np.log(fit_df["revenue"].values)
coef, _, _, _ = np.linalg.lstsq(X, y_log, rcond=None)

elasticity = pd.Series(coef[1:], index=channels, name="elasticity")
print("Elasticities (beta):")
print(elasticity)

# Compute marginal ROI per row: mROI = beta * (revenue / spend).
for c in channels:
    model_df[f"mroi_{c.replace('spend_', '')}"] = np.where(
        model_df[c] > 0,
        elasticity[c] * (model_df["revenue"] / model_df[c]),
        np.nan,
    )

mroi_summary = pd.DataFrame({
    "channel": channels,
    "elasticity": [elasticity[c] for c in channels],
    "avg_mROI": [model_df[f"mroi_{c.replace('spend_', '')}"].mean(skipna=True) for c in channels],
    "median_mROI": [model_df[f"mroi_{c.replace('spend_', '')}"].median(skipna=True) for c in channels],
})

mroi_summary
```

```python
# Quick fix: log1p model keeps zero-spend and zero-revenue rows.
channels = ["spend_google", "spend_meta", "spend_tiktok"]

fit_df_log1p = model_df[["revenue"] + channels].copy()

# Only impossible values are negatives; zeros are allowed with log1p.
valid_log1p = (fit_df_log1p["revenue"] >= 0)
for c in channels:
    valid_log1p &= (fit_df_log1p[c] >= 0)

fit_df_log1p = fit_df_log1p.loc[valid_log1p].copy()
dropped_log1p = len(model_df) - len(fit_df_log1p)
print(f"Rows used for log1p fit: {len(fit_df_log1p):,} | Dropped (negative values): {dropped_log1p:,}")

# log(revenue + 1) ~ log(spend + 1)
X_log1p = np.column_stack([
    np.ones(len(fit_df_log1p))
    ] + [np.log1p(fit_df_log1p[c].values) for c in channels]
 )
y_log1p = np.log1p(fit_df_log1p["revenue"].values)
coef_log1p, _, _, _ = np.linalg.lstsq(X_log1p, y_log1p, rcond=None)

elasticity_log1p = pd.Series(coef_log1p[1:], index=channels, name="elasticity_log1p")
print("Log1p elasticities (beta):")
print(elasticity_log1p)

# mROI approximation using beta * (revenue / spend), undefined where spend==0.
for c in channels:
    model_df[f"mroi_log1p_{c.replace('spend_', '')}"] = np.where(
        model_df[c] > 0,
        elasticity_log1p[c] * (model_df["revenue"] / model_df[c]),
        np.nan,
    )

mroi_log1p_summary = pd.DataFrame({
    "channel": channels,
    "elasticity_log1p": [elasticity_log1p[c] for c in channels],
    "avg_mROI_log1p": [model_df[f"mroi_log1p_{c.replace('spend_', '')}"].mean(skipna=True) for c in channels],
    "median_mROI_log1p": [model_df[f"mroi_log1p_{c.replace('spend_', '')}"].median(skipna=True) for c in channels],
})

mroi_log1p_summary
```

```python
# Quick fixes: time controls, lag features, weekly aggregation, and correlation sanity check.
work_df = model_df.copy()
work_df["date"] = pd.to_datetime(work_df["date"])

work_df = work_df.sort_values(["brand_id", "date"]).reset_index(drop=True)

# 1) Time controls
work_df["time_index"] = (work_df["date"] - work_df["date"].min()).dt.days
work_df["day_of_week"] = work_df["date"].dt.dayofweek
work_df["month"] = work_df["date"].dt.month

# C(day_of_week) and C(month) via one-hot encoding (drop_first avoids full collinearity).
dow_dummies = pd.get_dummies(work_df["day_of_week"], prefix="dow", drop_first=True, dtype=int)
month_dummies = pd.get_dummies(work_df["month"], prefix="month", drop_first=True, dtype=int)

# 2) Adstock-lite lags (within brand).
work_df["meta_lag1"] = work_df.groupby("brand_id")["spend_meta"].shift(1).fillna(0.0)
work_df["google_lag1"] = work_df.groupby("brand_id")["spend_google"].shift(1).fillna(0.0)

# Feature table for modeling.
model_features = pd.concat([
    work_df[[
        "date",
        "brand_id",
        "revenue",
        "spend_google",
        "spend_meta",
        "spend_tiktok",
        "google_lag1",
        "meta_lag1",
        "time_index",
        "day_of_week",
        "month",
    ]],
    dow_dummies,
    month_dummies,
], axis=1)

# 3) Weekly aggregation (less noisy than daily).
work_df["week"] = work_df["date"].dt.to_period("W").dt.start_time
weekly_df = (
    work_df.groupby(["week", "brand_id"], as_index=False)[
        ["revenue", "spend_google", "spend_meta", "spend_tiktok", "google_lag1", "meta_lag1"]
    ]
    .sum()
    .sort_values(["brand_id", "week"])
    .reset_index(drop=True)
)
weekly_df["time_index"] = ((weekly_df["week"] - weekly_df["week"].min()).dt.days // 7).astype(int)
weekly_df["day_of_week"] = weekly_df["week"].dt.dayofweek
weekly_df["month"] = weekly_df["week"].dt.month

# 4) Correlation sanity checks.
corr_daily = work_df[["revenue", "spend_meta", "spend_google", "spend_tiktok"]].corr(numeric_only=True)
corr_weekly = weekly_df[["revenue", "spend_meta", "spend_google", "spend_tiktok"]].corr(numeric_only=True)

print("Built model_features with time controls + lag features.")
print(f"Daily rows: {len(work_df):,} | Weekly rows: {len(weekly_df):,}")
print("Suggested formula:")
print("np.log1p(revenue) ~ np.log1p(spend_google) + np.log1p(spend_meta) + np.log1p(spend_tiktok) + np.log1p(google_lag1) + np.log1p(meta_lag1) + time_index + C(day_of_week) + C(month)")

print("\nDaily correlation matrix:")
display(corr_daily)
print("\nWeekly correlation matrix:")
display(corr_weekly)

weekly_df.head()
```

```python
# Collinearity fix: combine Google + Meta into one performance channel.
combo_df = model_df.copy()

combo_df["performance_spend"] = combo_df["spend_google"] + combo_df["spend_meta"]

# Keep zeros with log1p; drop only impossible negatives.
fit_combo = combo_df[["revenue", "performance_spend", "spend_tiktok"]].copy()
valid_combo = (
    (fit_combo["revenue"] >= 0)
    & (fit_combo["performance_spend"] >= 0)
    & (fit_combo["spend_tiktok"] >= 0)
)
fit_combo = fit_combo.loc[valid_combo].copy()

print(f"Rows used (combined-channel model): {len(fit_combo):,} / {len(combo_df):,}")

# log1p(revenue) ~ log1p(performance_spend) + log1p(tiktok)
X_combo = np.column_stack([
    np.ones(len(fit_combo)),
    np.log1p(fit_combo["performance_spend"].values),
    np.log1p(fit_combo["spend_tiktok"].values),
])
y_combo = np.log1p(fit_combo["revenue"].values)
coef_combo, _, _, _ = np.linalg.lstsq(X_combo, y_combo, rcond=None)

elasticity_combo = pd.Series(
    {
        "performance_spend": coef_combo[1],
        "spend_tiktok": coef_combo[2],
    },
    name="elasticity_log1p",
)
print("Combined-channel elasticities (log1p model):")
print(elasticity_combo)

# mROI approximation: beta * (revenue / spend), undefined where spend == 0.
combo_df["mroi_performance"] = np.where(
    combo_df["performance_spend"] > 0,
    elasticity_combo["performance_spend"] * (combo_df["revenue"] / combo_df["performance_spend"]),
    np.nan,
)
combo_df["mroi_tiktok_combo"] = np.where(
    combo_df["spend_tiktok"] > 0,
    elasticity_combo["spend_tiktok"] * (combo_df["revenue"] / combo_df["spend_tiktok"]),
    np.nan,
)

combo_mroi_summary = pd.DataFrame({
    "channel": ["performance_spend", "spend_tiktok"],
    "elasticity_log1p": [
        elasticity_combo["performance_spend"],
        elasticity_combo["spend_tiktok"],
    ],
    "avg_mROI": [
        combo_df["mroi_performance"].mean(skipna=True),
        combo_df["mroi_tiktok_combo"].mean(skipna=True),
    ],
    "median_mROI": [
        combo_df["mroi_performance"].median(skipna=True),
        combo_df["mroi_tiktok_combo"].median(skipna=True),
    ],
})

combo_mroi_summary
```

```python
# Sequential modeling on weekly data (recommended for decisions).

# Build weekly data if the earlier quick-fix cell was not run.
if "weekly_df" not in globals():
    temp = model_df.copy()
    temp["date"] = pd.to_datetime(temp["date"])
    temp["week"] = temp["date"].dt.to_period("W").dt.start_time
    weekly_df = (
        temp.groupby(["week", "brand_id"], as_index=False)[
            ["revenue", "spend_google", "spend_meta", "spend_tiktok"]
        ]
        .sum()
        .sort_values(["brand_id", "week"])
        .reset_index(drop=True)
    )

# Step 1: Model total performance at weekly level.
w = weekly_df.copy()
for c in ["revenue", "spend_google", "spend_meta", "spend_tiktok"]:
    w[c] = pd.to_numeric(w[c], errors="coerce")

valid = (w["revenue"] >= 0) & (w["spend_google"] >= 0) & (w["spend_meta"] >= 0) & (w["spend_tiktok"] >= 0)
w_fit = w.loc[valid].copy()
w_fit["performance_spend"] = w_fit["spend_google"] + w_fit["spend_meta"]

X_seq = np.column_stack([
    np.ones(len(w_fit)),
    np.log1p(w_fit["performance_spend"].values),
    np.log1p(w_fit["spend_tiktok"].values),
])
y_seq = np.log1p(w_fit["revenue"].values)
coef_seq, _, _, _ = np.linalg.lstsq(X_seq, y_seq, rcond=None)

weekly_elasticity = pd.Series(
    {
        "performance_spend": coef_seq[1],
        "spend_tiktok": coef_seq[2],
    },
    name="weekly_elasticity_log1p",
)

print("Weekly sequential model fitted (recommended decision view).")
print("Step 1 model: log1p(revenue) ~ log1p(performance_spend) + log1p(spend_tiktok)")
print(weekly_elasticity)

# Step 2: Meta vs Google split using business efficiency metrics when available.
# Expected optional columns in model_df:
#  - cac_google, cac_meta
#  - conv_rate_google, conv_rate_meta
#  - click_eff_google, click_eff_meta

split_inputs = [
    "cac_google", "cac_meta",
    "conv_rate_google", "conv_rate_meta",
    "click_eff_google", "click_eff_meta",
]

available_split_cols = [c for c in split_inputs if c in model_df.columns]
if len(available_split_cols) == len(split_inputs):
    split_df = model_df[split_inputs].copy()
    split_df = split_df.apply(pd.to_numeric, errors="coerce")
    split_df = split_df.dropna()

    if len(split_df) > 0:
        # Higher score means expected better marginal efficiency.
        google_score = (split_df["conv_rate_google"] * split_df["click_eff_google"]) / split_df["cac_google"].replace(0, np.nan)
        meta_score = (split_df["conv_rate_meta"] * split_df["click_eff_meta"]) / split_df["cac_meta"].replace(0, np.nan)

        score_df = pd.DataFrame({"google_score": google_score, "meta_score": meta_score}).replace([np.inf, -np.inf], np.nan).dropna()
        google_share = (score_df["google_score"] / (score_df["google_score"] + score_df["meta_score"])).mean()
        meta_share = 1 - google_share

        split_summary = pd.DataFrame({
            "channel": ["google", "meta"],
            "avg_efficiency_share": [google_share, meta_share],
        })
        print("\nStep 2 split (from CAC + conversion + click efficiency):")
        display(split_summary)
    else:
        print("\nStep 2 split skipped: split columns exist but have no complete rows.")
else:
    missing_split_cols = [c for c in split_inputs if c not in model_df.columns]
    print("\nStep 2 split skipped: missing business metrics columns.")
    print(f"Missing: {missing_split_cols}")

print("\nWeekly is preferred for decisions: smoother, less noisy, and usually more stable coefficients.")
```

```python
# 1. Diminishing Returns Check (CRITICAL)
# Plot: spend vs mROI per channel
# If you DON'T see a downward slope → your data is biased

import matplotlib.pyplot as plt
from scipy import stats as sp_stats

channels = [
    ("spend_google", "mroi_log1p_google", "Google"),
    ("spend_meta", "mroi_log1p_meta", "Meta"),
    ("spend_tiktok", "mroi_log1p_tiktok", "TikTok"),
]

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
diminishing_results = []

for ax, (spend_col, mroi_col, label) in zip(axes, channels):
    # Prepare data
    d = model_df[[spend_col, mroi_col]].copy()
    d = d.replace([np.inf, -np.inf], np.nan).dropna()
    d = d[d[spend_col] > 0].sort_values(spend_col)
    
    if d.empty:
        ax.text(0.5, 0.5, f"{label}\nNo valid data", ha="center", va="center", transform=ax.transAxes)
        ax.set_title(f"{label} (No Data)")
        continue
    
    # Scatter plot
    ax.scatter(d[spend_col], d[mroi_col], alpha=0.15, s=20, color="#4C78A8")
    
    # Fit trend line (linear regression)
    slope, intercept, r_value, p_value, std_err = sp_stats.linregress(d[spend_col], d[mroi_col])
    line_x = np.array([d[spend_col].min(), d[spend_col].max()])
    line_y = slope * line_x + intercept
    
    # Plot trend line
    color = "#2A9D8F" if slope < 0 else "#E45756"  # Green if downward, red if upward/flat
    ax.plot(line_x, line_y, color=color, linewidth=2.5, label=f"Trend (slope={slope:.4f})")
    
    # Formatting
    ax.set_xlabel("Spend ($)")
    ax.set_ylabel("mROI")
    ax.set_title(f"{label}\nSlope: {slope:.4f} | R²: {r_value**2:.3f}")
    ax.legend(loc="best", fontsize=9)
    ax.grid(True, alpha=0.2)
    
    # Store results
    has_diminishing = slope < 0
    diminishing_results.append({
        "channel": label,
        "slope": slope,
        "r_squared": r_value**2,
        "p_value": p_value,
        "diminishing_returns": "✅ Yes" if has_diminishing else "❌ No (Suspicious)",
        "interpretation": "Healthy - mROI decreases as spend increases" if has_diminishing else "⚠️ WARNING: Flat/upward slope suggests bias or insufficient spend variation",
    })

plt.suptitle("📊 Diminishing Returns Check: Spend vs mROI per Channel\n(Downward slope = healthy diminishing returns)", fontsize=12, fontweight="bold")
plt.tight_layout()
plt.show()

# Summary
print("=" * 80)
print("🔍 DIMINISHING RETURNS ANALYSIS")
print("=" * 80)

for result in diminishing_results:
    print(f"\n{result['channel']}:")
    print(f"  Slope: {result['slope']:.6f}")
    print(f"  R²: {result['r_squared']:.4f}")
    print(f"  Diminishing Returns: {result['diminishing_returns']}")
    print(f"  → {result['interpretation']}")

print("\n" + "=" * 80)
print("⚠️ CRITICAL CHECK:")
if all(r["diminishing_returns"] == "✅ Yes" for r in diminishing_results):
    print("✅ All channels show healthy diminishing returns patterns.")
    print("   Your mROI estimates are likely RELIABLE.")
else:
    print("❌ Some channels lack clear diminishing returns!")
    print("   Possible causes:")
    print("   • Data contains outliers or extreme values")
    print("   • Spend is too concentrated in narrow range")
    print("   • Model misspecification or data quality issues")
    print("   • True relationship is non-linear or threshold-based")
    print("\n   Recommendation: Review data quality & consider domain knowledge")
print("=" * 80)

# Display detailed results
diminishing_df = pd.DataFrame(diminishing_results)
display(diminishing_df)
```

```python
# 🚀 If you want to level this up: joint-effects modeling to reduce attribution bias

# Robust setup: allow zeros with log1p, remove negatives only.
joint_df = model_df[["revenue", "spend_google", "spend_meta", "spend_tiktok"]].copy()
for c in ["revenue", "spend_google", "spend_meta", "spend_tiktok"]:
    joint_df[c] = pd.to_numeric(joint_df[c], errors="coerce")

joint_df = joint_df.dropna()
joint_df = joint_df[(joint_df["revenue"] >= 0) & (joint_df["spend_google"] >= 0) & (joint_df["spend_meta"] >= 0)]

if len(joint_df) < 30:
    raise ValueError("Too few rows for stable joint-effects model.")

# Base model: log(revenue) ~ log(google) + log(meta)
# Using log1p to keep zero-spend rows.
X_base = np.column_stack([
    np.ones(len(joint_df)),
    np.log1p(joint_df["spend_google"].values),
    np.log1p(joint_df["spend_meta"].values),
])
y_base = np.log1p(joint_df["revenue"].values)
coef_base, _, _, _ = np.linalg.lstsq(X_base, y_base, rcond=None)
pred_base = X_base @ coef_base
r2_base = 1 - np.sum((y_base - pred_base) ** 2) / np.sum((y_base - y_base.mean()) ** 2)

# Joint-effects model with interaction: google * meta lift
joint_df["log_google"] = np.log1p(joint_df["spend_google"])
joint_df["log_meta"] = np.log1p(joint_df["spend_meta"])
joint_df["google_meta_interaction"] = joint_df["log_google"] * joint_df["log_meta"]

X_inter = np.column_stack([
    np.ones(len(joint_df)),
    joint_df["log_google"].values,
    joint_df["log_meta"].values,
    joint_df["google_meta_interaction"].values,
])
y_inter = y_base.copy()
coef_inter, _, _, _ = np.linalg.lstsq(X_inter, y_inter, rcond=None)
pred_inter = X_inter @ coef_inter
r2_inter = 1 - np.sum((y_inter - pred_inter) ** 2) / np.sum((y_inter - y_inter.mean()) ** 2)

summary = pd.DataFrame({
    "model": [
        "Base: log1p(revenue) ~ log1p(google) + log1p(meta)",
        "Joint: + log1p(google)*log1p(meta)",
    ],
    "google_coef": [coef_base[1], coef_inter[1]],
    "meta_coef": [coef_base[2], coef_inter[2]],
    "interaction_coef": [np.nan, coef_inter[3]],
    "r_squared": [r2_base, r2_inter],
})

print("Rows used:", f"{len(joint_df):,}")
print("\nModel comparison:")
display(summary)

print("Interpretation:")
if coef_inter[3] > 0:
    print("- Positive interaction: Google and Meta likely reinforce each other (cross-channel lift).")
elif coef_inter[3] < 0:
    print("- Negative interaction: channels may cannibalize each other at higher spend levels.")
else:
    print("- Near-zero interaction: limited evidence of cross-channel lift in this specification.")

print("- As joint effects are modeled, single-channel attribution bias is reduced vs one-channel-at-a-time views.")
```
